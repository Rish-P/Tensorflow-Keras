```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

datadir="/Users/veer/Downloads/animals"
training_data = []
def create_training_data():
  path = os.path.join(datadir, "train")
  for item in os.listdir(path):
    img = cv2.imread(os.path.join(path, item), cv2.IMREAD_GRAYSCALE)
    final = cv2.resize(img, (80,80))
    training_data.append([final, item[:3]]) # getting the first 3 words, viz cat or dog

create_training_data()

for item in training_data[:10]:
  print(item)

X=[]
y=[]

for features, label in training_data:
  X.append(features)
  y.append(label)

X = np.array(X).reshape(-1, 80, 80, 1)







#for item in training_data[:10]:
  #print(item)
```
